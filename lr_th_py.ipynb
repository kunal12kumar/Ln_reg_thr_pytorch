```python
import pandas as pd 
import numpy as np 
import torch.nn as nn
import torch 
import torch.optim as optim
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder , StandardScaler , OneHotEncoder
```

```python
# reading the data
ins=pd.read_csv('insurance.csv')
ins.head(5)
```

```python
ins.describe()
```

```python
ins.info()
```

```python
# there are three object value first convert it into the numerical value by encoding 
print(ins['sex'].value_counts() )
print("\n")
print(ins['smoker'].value_counts())
print("\n")
print(ins['region'].value_counts())
```

```python
#from above we can see that sex and smoker has binary therefore we will use Label encoder for this but for the region we need to use one hot Encoder
label_encoder=LabelEncoder()
ins['sex']=label_encoder.fit_transform(ins['sex'])
```

```python
ins['sex']
```

```python
# now do same for the smoker
ins['smoker']=label_encoder.fit_transform(ins['smoker'])
ins['smoker']
```

```python
ins.head(5)
```

```python
# now for the region we will do one hot encoding 
one_hot=OneHotEncoder()
one_hot_region=one_hot.fit_transform(ins[['region']])
one_hot_region_feature_names=one_hot.get_feature_names_out(['region'])
one_hot_region_feature_names
# need to convert it into the array other wise it will a float value
one_hot_region.shape
print(one_hot_region_feature_names.shape)
print(one_hot_region.shape)
```

```python
# now form a data fram
one_hot_region_df=pd.DataFrame(one_hot_region.toarray() , columns=one_hot_region_feature_names)
```

```python
one_hot_region_df.head()
```

```python
# now merge both the dataframe original one and this one 
updated_data=pd.concat([ins.drop('region' ,axis=1) , one_hot_region_df ])
updated_data.head(5)
```
